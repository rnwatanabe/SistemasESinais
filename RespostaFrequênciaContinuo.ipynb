## Definição

Aqui, vamos analisar qual a resposta de um sistema contínuo linear e invariante no tempo quando a entrada é uma exponencial complexa.

Vamos considerar uma entrada do tipo:

$$x(t) = e^{j\omega t}$$

Considerando que um sistema tem a resposta ao impulso $h(t)$, a saída desse sistema (estado nulo) para a entrada $x(t)$ é:

$$y(t) = \int_{-\infty}^\infty h(\tau)e^{j\omega(t-\tau)\,d\tau} = e^{j\omega t}\underbrace{\int_{-\infty}^\infty h(\tau)e^{-j\omega\tau\,d\tau}}_{H(j\omega)} = H(j\omega)e^{j\omega t}$$

A função $H(j\omega)$ é conhecida como **Função Resposta em Frequência**.  A expressão anterior mostra que se colocarmos uma exponencial complexa de qualquer frequência $\omega$, a saída será uma exponencial complexa com a mesma frequência $\omega$, multiplicada por $H(j\omega)$.  Isso acontece porque a função $e^{j\omega t}$ é uma **autofunção** de qualquer sistema linear e invariantre no tempo.


Note que, como a resposta em frequência é calculada a partir da resposta ao impulso, a resposta em frequência também caracteriza completamente o comportamento de um sistema linear e invariante no tempo. Isso significa que se conhecermos a função resposta em frequência de um sistema linear e invariante no tempo, conhecemos completamente o seu comportamento.  Por isso, uma outra maneira de representar um sistema em um diagrama de blocos é com a sua resposta em frequência.

Até agora, já conhecemos várias formas para caracterizar um sistema linear e invariante no tempo e sempre podemos uma representação a partir de outra:

- Equação diferencial

### Exemplo 1

Vamos encontrar a resposta em frequência de um sistema que tenha a resposta ao impulso:

$$h(t) = e^{-2t}u(t)$$

A função resposta em frequência desse sistema é:

$$H(j\omega) = \int_{-\infty}^\infty h(\tau)e^{-j\omega\tau}\,d\tau = \int_{-\infty}^\infty e^{-2\tau}u(\tau)e^{-j\omega\tau}\,d\tau = \int_{0}^\infty e^{-(2+j\omega)\tau}\,d\tau = \left.\frac{e^{-(2+j\omega)\tau}}{-(2+j\omega)}\right|_{\tau=0}^\infty=\frac{1}{2+j\omega}$$


---

## Existência da Resposta em frequência

A resposta em frequência só vai existir se a integral que define a resposta em frequência convergir para um número finito:

$|\int_{-\infty}^\infty h(t)e^{-j\omega t}\,dt| < \infty$

Para isso acontecer, temos que:

$|\int_{-\infty}^\infty h(t)e^{j\omega t}\,dt| \leq \int_{-\infty}^\infty |h(t)e^{j\omega t}|\,dt =  \int_{-\infty}^\infty |h(t)||e^{j\omega t}|\,dt = \int_{-\infty}^\infty |h(t)|\,dt < \infty$

Ou seja, para a resposta em frequência de um sistema existir, temos que $\int_{-\infty}^\infty |h(t)|\,dt < \infty$, que é exatamente o critério de estabilidade de um sistema linear e invariante no tempo. Portanto, **um sistema tem uma resposta em frequência apenas se o sistema for estável**.

## Ganho e fase

O valor de $H(j\omega)$ é um número complexo que pode ser escrito na forma polar:

$$H(j\omega) = |H(j\omega)|e^{j\phi(\omega)}$$

O $|H(j\omega)|$ é o módulo de $H(j\omega)$, e muitas vezes é chamado de **ganho** do sistema. O termo $\phi(\omega)$ é a fase de $H(j\omega)$ e muitas é conhecido como **deslocamento de fase** do sistema.

Então, quando um sinal $x(t) = e^{j\omega t}$ é usado como entrada, a sua saída (estado nulo) será:

$$y(t) = H(j\omega)e^{j\omega t} = |H(j\omega)|e^{j\phi(\omega)}e^{j\omega t} = |H(j\omega)|e^{j(\omega t+\phi(\omega))}$$

Quando o deslocamento de fase é zero, dizemos que a entrada em saída estão **em fase**.

Analisando a expressão da função de resposta em frequência, chegamos em duas propriedades (fica como exercício a prova):

- $|H(j\omega)| = |H(-j\omega)|$

- $\phi(\omega) = - \phi(-\omega)$


___
### Exemplo 2

Considerando o exemplo anterior, vamos calcular o ganho e o deslocamento de fase da resposta em frequência. A resposta em frequência é:

$$H(j\omega)=\frac{1}{2+j\omega}$$

O ganho é:

$$|H(j\omega)| = \left|\frac{1}{2+j\omega} \right| = \frac{1}{|2+j\omega|} = \frac{1}{\sqrt{4+\omega^2}}$$

e o deslocamento de fase é:

$$\phi(\omega) = -\arctan\left(\frac{\omega}{2}\right)$$

### Exemplo 3

Considerando o ganho e deslocamento de fase do exemplo anterior, vamos encontrar o sinal de saída (considerando estado inicial nulo) do sistema se for usado como sinal de entrada $x(t) = e^{j2t}$.

Nesse caso, $\omega=2$ rad/s. O ganho e deslocamento de fase para essa frequência angular é:

$$|H(j2)|  = \frac{1}{\sqrt{4+2^2}}=\frac{1}{4}$$

$$\phi(2) = -\arctan\left(\frac{2}{2}\right) = -\arctan\left(1\right) = -\frac{\pi}{4}$$


Então, a saída do sistema é:

$$y(t) = |H(j2)|e^{j(2t+\phi(2))} = \frac{1}{4}e^{j\left(2t-\frac{\pi}{4}\right)} $$

## Resposta em frequência a partir de uma equação diferencial

Considerando o formato geral de uma equação diferencial de um sistema linear e invariante no tempo:

$$\frac{d^Ny}{dt^N} + a_{N-1}\frac{d^{N-1}y}{dt^{N-1}}+...+a_0y(t) = b_M\frac{d^Mx}{dt^M} +...+b_2\frac{d^2x}{dt^2}+b_1\frac{dx}{dt} + b_0x(t) $$

Se $x(t) = e^{j\omega t}$, a sua saída será $y(t) = H(j\omega)e^{j\omega t}$. Então:

$$(j\omega)^NH(j\omega)e^{j\omega t} + a_{N-1}(j\omega)^{N-1}H(j\omega)e^{j\omega t}+...+a_0H(j\omega)e^{j\omega t} = b_M(j\omega)^Me^{j\omega t} +...+b_2(j\omega)^2e^{j\omega t}+b_1(j\omega)e^{j\omega t} + b_0e^{j\omega t} $$

Como $e^{j\omega t}$ nunca é igual a zero, podemos fazer:

$$(j\omega)^NH(j\omega) + a_{N-1}(j\omega)^{N-1}H(j\omega)+...+a_0H(j\omega) = b_M(j\omega)^M +...+b_2(j\omega)^2+b_1(j\omega) + b_0 $$

Então, a função resposta em frequência é:


$$H(j\omega) = \frac{b_M(j\omega)^M +...+b_2(j\omega)^2+b_1(j\omega) + b_0}{(j\omega)^N + a_{N-1}(j\omega)^{N-1}+...+a_0}$$

___
### Exemplo 4

Considerando um sistema com a seguinte equação diferencial:

$$\frac{d^2y}{dt^2} + 4\frac{dy}{dt} + 5y(t) = \frac{dx}{dt} + 2x(t)$$

Então a resposta em frequência desse sistema é:

$$H(j\omega) = \frac{j\omega + 2}{(j\omega)^2+4j\omega+5} = \frac{j\omega + 2}{5-\omega^2+4j\omega}$$

## Resposta em frequência a partir da função de transferência

Para sistemas causais e estáveis, a resposta em frequência pode ser obtida da função de transferência:

$H(j\omega) = \left.H(s)\right|_{s=j\omega}$

Para sistemas não causais, a função de transferência não é definida (estamos usando a transformada de Laplace unilateral), mas a resposta em frequência é.

---
### Exemplo 5

Um sistema linear e invariante no tempo tem a seguinte resposta ao impulso:

$h(t) = e^{-0.5t}u(t)$

A Função de transferência desse sistema é:

$H(s) = \frac{1}{s+0.5}$

Então, a resposta em frequência desse sistema é:

$H(j\omega) = \frac{1}{j\omega+0.5}$

## Resposta de entradas senoidais

Vamos considerar um sistema linear e invariante no tempo com uma resposta em frequência $H(j\omega)$ e uma entrada $x(t) = \cos(\omega t+\theta)$.  Primeiro, vamos reescrever o sinal de entrada como uma soma de duas exponenciais complexas.

$x(t) = \cos(\omega t+\theta) = \frac{e^{j(\omega t + \theta)} + e^{-j(\omega t + \theta)}}{2} = \frac{e^{j\omega t}e^{j\theta} + e^{-j\omega t}e^{-j\theta}}{2} $


Então, a saída do sistema para essa entrada é:

$y(t) = \frac{|H(j\omega)|e^{j(\omega t+\phi(\omega))}e^{j\theta} + |H(-j\omega)|e^{-j\omega t+j\phi(-\omega)}e^{-j\theta}}{2} = \frac{|H(j\omega)|e^{j(\omega t+\theta+\phi(\omega))} + |H(j\omega)|e^{-j(\omega t+\theta+\phi(\omega))}}{2} = |H(j\omega)| \frac{e^{j(\omega t+\theta+\phi(\omega))} + e^{-j(\omega t+\theta+\phi(\omega))}}{2} =  |H(j\omega)|\cos(\omega t+\theta+\phi(\omega))$

## Gráfico da resposta em frequência

Normalmente fazemos dois gráficos para representar a resposta em frequência: um do **ganho (módulo)** e outro da **fase**.

### Exemplo 7

Vamos considerar o sistema do exemplo 1

$H(j\omega) = \frac{1}{j\omega+2}$

O módulo é:

$|H(j\omega)| = \frac{1}{\sqrt{4+\omega^2}}$

e a fase:

$\angle{H(j\omega)} = -\arctan\left(\frac{\omega}{2}\right)$

O gráfico do módulo e da fase são mostrados abaixo:


---

```python
omega = sym.symbols('\omega')
Hganho = 1/sym.sqrt(4+omega**2)
Hfase = -sym.atan(omega/2)
p1 = sym.plot(Hganho, (omega, -100, 100),show=False, title='$|H(j\omega)|$')
p2 = sym.plot(Hfase, (omega, -100, 100),show=False, title=r'$\angle H(j\omega)$')
PlotGrid(2,1,p1,p2);
```

Perceba que a informação contida para frequências negativas é a mesma que está no gráfico para frequências positivas. Então é muito comum mostrar apenas o lado com frequências positivas.

```python
omega = sym.symbols('\omega')
Hganho = 1/sym.sqrt(4+omega**2)
Hfase = -sym.atan(omega/2)
p1 = sym.plot(Hganho, (omega, 0, 100),show=False, title='$|H(j\omega)|$')
p2 = sym.plot(Hfase, (omega, 0, 100),show=False, title=r'$\angle H(j\omega)$')
PlotGrid(2,1,p1,p2);
```

Outra coisa muito comum é mostrar o ganho na unidade **dB (decibel)**. Isso é feito fazendo o logaritmo da magnitude e multiplicando por 20.

Em db, a magnitude da resposta em frequência fica:

$|H(j\omega)| = 20\log\left(\frac{1}{|j\omega+2|}\right)= 20\log(1)-20\log(|j\omega+2|)=-20\log(|j\omega+2|) = -20\log(\sqrt{\omega^2+4}) = -10\log(\omega^2+4)$

```python
omega = sym.symbols('\omega')
Hganho = 1/sym.sqrt(4+omega**2)
Hfase = -sym.atan(omega/2)
p1 = sym.plot(-10*sym.log(omega**2+4,10), (omega, 0, 100),show=False, title='$|H(j\omega)|$ (dB)')
p2 = sym.plot(Hfase, (omega, 0, 100),show=False, title=r'$\angle H(j\omega)$')
PlotGrid(2,1,p1,p2);
```

Outra forma comum de mostrar a resposta em frequência  é representando a frequência em escala logarítmica

```python
omega = sym.symbols('\omega')
Hganho = 1/sym.sqrt(4+omega**2)
Hfase = -sym.atan(omega/2)
p1 = sym.plot(-10*sym.log(omega**2+4,10), (omega, 0.01, 100),show=False, title='$|H(j\omega)|$ (dB)', xscale='log')
p2 = sym.plot(Hfase, (omega, 0.01, 100),show=False, title=r'$\angle H(j\omega)$', xscale='log')
PlotGrid(2,1,p1,p2);
```

Uma outra forma de  se mostrar a resposta em frequência  é mostrando a fase em graus.

```python
omega = sym.symbols('\omega')
Hganho = 1/sym.sqrt(4+omega**2)
Hfase = -sym.atan(omega/2)*180/np.pi
p1 = sym.plot(-10*sym.log(omega**2+4,10), (omega, 0.01, 100),show=False, title='$|H(j\omega)|$ (dB)', xscale='log')
p2 = sym.plot(Hfase, (omega, 0.01, 100),show=False, title=r'$\angle H(j\omega)$', xscale='log')
PlotGrid(2,1,p1,p2);
```

### Diagrama de Bode

Quando a resposta em frequência está na forma racional, existe uma maneira de fazer uma <u>aproximação</u> do gráfico da resposta em frequência utilizando retas assíntotas, conhecido como **diagrama de Bode**. Nesse método o gráfico do ganho é feito em dB e a frequência em escala logarítmica.

Nesta situação, a resposta em frequência pode ser colocada na forma de polos e zeros:

$H(j\omega) = \frac{K(j\omega - z_1)(j\omega-z_2)...(j\omega-z_M)}{(j\omega - p_1)(j\omega-p_2)...(j\omega-p_N)}$

Para facilitar a análise posterior, vamos multiplicar e dividir cada um dos fatores pelo seu respectivo zero ou polo:

$ H(j\omega) = \frac{Kz_1z_2...z_M}{p_1p_2...p_N}\frac{\left(j\frac{\omega}{z_1} - 1\right)\left(j\frac{\omega}{z_2}-1\right)...\left(j\frac{\omega}{z_M}-1\right)}{\left(j\frac{\omega}{p_1} - 1\right)\left(j\frac{\omega}{p_2}-1\right)...\left(j\frac{\omega}{p_N}-1\right)} $

#### Ganho

Calculando o ganho  dessa função, temos:

$|H(j\omega)| = \left|\frac{Kz_1z_2...z_M}{p_1p_2...p_N}\right|\frac{\left|j\frac{\omega}{z_1} - 1\right|\left|j\frac{\omega}{z_2}-1\right|...\left|j\frac{\omega}{z_M}-1\right|}{\left|j\frac{\omega}{p_1} - 1\right|\left|j\frac{\omega}{p_2}-1\right|...\left|j\frac{\omega}{p_N}-1\right|}$


Fazendo esse ganho em dB:

$20\log_{10}(|H(j\omega)|) = 20\log_{10}\left(\left|\frac{Kz_1z_2...z_M}{p_1p_2...p_N}\right|\frac{\left|j\frac{\omega}{z_1} - 1\right|\left|j\frac{\omega}{z_2}-1\right|...\left|j\frac{\omega}{z_M}-1\right|}{\left|j\frac{\omega}{p_1} - 1\right|\left|j\frac{\omega}{p_2}-1\right|...\left|j\frac{\omega}{p_N}-1\right|}\right)$

Lembrando que a multiplicação de logaritmos viram somas e divisão viram subtrações:

$20\log_{10}(|H(j\omega)|) = 20\log_{10}\left(\left|\frac{Kz_1z_2...z_M}{p_1p_2...p_N}\right|\right)+20\log_{10}\left(\left|j\frac{\omega}{z_1} - 1\right|\right)+20\log_{10}\left(\left|j\frac{\omega}{z_2}-1\right|\right)+...+20\log_{10}\left(\left|j\frac{\omega}{z_M}-1\right|\right)-20\log_{10}\left(\left|j\frac{\omega}{p_1} - 1\right|\right)-20\log_{10}\left(\left|j\frac{\omega}{p_2}-1\right|\right)-...-20\log_{10}\left(\left|j\frac{\omega}{p_N}-1\right|\right)$


Vamos analisar apenas o primeiro desse termos:

$20\log_{10}\left(\left|\frac{Kz_1z_2...z_M}{p_1p_2...p_N}\right|\right)$

Esse é o ganho do sistema em $\omega=0 $ rad/s.

- **Termos correspondentes a zeros**

Agora, vamos analisar um dos outros termos, que são iguais na sua forma:

$20\log_{10}\left(\left|j\frac{\omega}{z_1} - 1\right|\right)=20\log_{10}\left(\sqrt{\frac{\omega^2}{z_1^2} + 1}\right)=10\log_{10}\left(\frac{\omega^2}{z_1^2} + 1\right)$

Para $\omega=0$, o valor da expressão acima é $10\log_{10}(1)=0$. Para $\omega\rightarrow\infty$, o o termo com $\omega$ domina a expresão e fica:

$10\log_{10}\left(\frac{\omega^2}{z_1^2} + 1\right)\approx 10\log_{10}\left(\frac{\omega^2}{z_1^2}\right)=20\log_{10}\left(\frac{\omega}{z_1}\right)$

Ou seja, o gráfico devido a essa parte da expressão é zero perto de $\omega=0$ e uma reta que sobe 20 dB a cada década da frequência (multiplos de 10).

Quando $\omega=z_1$, o valor da expressão é $10\log_{10}(2)=3 dB$. Esse valor ainda é bem perto do zero e para a aproximação do diagrama de Bode, considera-se esse ponto como o valor que vai começar a reta para cima.

Abaixo está mostrado o gráfico, em azul, do diagrama de Bode. Em laranja, é mostrado o gráfico sem nenhuma aproximação.

```python
omega = sym.symbols('\omega')
z1 = 2
Happroxz = sym.Piecewise((0, ((omega>=0) & (omega<z1))),
                         (20*sym.log(omega/z1,10), (omega>=z1)))

p1 = sym.plot(Happroxz, (omega, 0.1, 100), xscale='log', title='$(s+z)$', show=False)
p2= sym.plot(10*sym.log(omega**2/z1**2+1,10),(omega, 0.1, 100), xscale='log', show=False,linestyle='--')

p1.extend(p2)
p1.show()
fig = p1._backend.fig
ax = p1._backend.ax[0]
ax.set_xticks([0.1, 1, 2, 10])
ax.set_xticklabels([r'$10^{-1}$', r'$10^{0}$', r'$z$', r'$10^1$']);
p1._backend.ax[0]._children[1].set_linestyle('dashed')
p1._backend.ax[0]._children[0].set_linewidth(7)
```

- **Termos correspondentes a polos**

Quando o termo se trata de um polo o valor da resposta em frequência em $\omega=0$ é 0, em $\omega=p_1$ é -3 dB e tendendo a infinito, se aproxima de $-20\log_{10}\left(\frac{\omega}{p_1}\right)$ (uma reta que cai a 20 db por década.)



```python
omega = sym.symbols('\omega')
polo1 = -2
Happroxz = sym.Piecewise((0, ((omega>=0) & (omega<-polo1))),
                         (-20*sym.log(-omega/polo1,10), (omega>=-polo1)))

p1 = sym.plot(Happroxz, (omega, 0.1, 100), xscale='log', title=r'$\frac{1}{s+p}$',show=False)
p2= sym.plot(-10*sym.log(omega**2/polo1**2+1,10),(omega, 0.1, 100), xscale='log', show=False,linestyle='--')

p1.extend(p2)
p1.show();
fig = p1._backend.fig
ax = p1._backend.ax[0]
ax.set_xticks([0.1, 1, 2, 10])
ax.set_xticklabels([r'$10^{-1}$', r'$10^{0}$', r'$p$', r'$10^1$']);
p1._backend.ax[0]._children[1].set_linestyle('dashed')
p1._backend.ax[0]._children[0].set_linewidth(7)
```

- **Termos correspondentes a zeros ou polos complexos conjugados**

Se por acaso os zeros ou polos contiverem valores complexos conjugados, os termos correspondentes a esses polos ou zeros será:

$20\log_{10}\left(\left|j\frac{\omega}{p_1} - 1\right|\left|j\frac{\omega}{p_1*}-1\right|\right)$

Podemos deixar esse termo como:

$20\log_{10}\left(\left|1-\frac{\omega^2}{\omega_r^2} +j2\xi\frac{\omega}{\omega_r}\right|\right) = 20\log_{10}\left[\sqrt{\left(1-\frac{\omega^2}{\omega_r^2}\right)^2 +4\xi^2\frac{\omega^2}{\omega_r^2}}\right]= 10\log_{10}\left[\left(1-\frac{\omega^2}{\omega_r^2}\right)^2 +4\xi^2\frac{\omega^2}{\omega_r^2}\right]$

Em que $xi$ é um número entre 0 e 1 e $\omega_r$ corresponde à frequência em que o valor desse ganho é máximo

Quando $\omega=0$, o termo acima é igual a zero.

Quando $\omega\rightarrow \infty$, o termo tende a:

$10\log_{10}\left[\frac{\omega^4}{\omega_r^4}\right]=40\log_{10}\left(\frac{\omega}{\omega_r}\right)$

Se o termo for correspondente a um zero, é uma reta subindo com inclinação de 40 dB a cada década de frequência. Se for um correspondente a um polo, será uma reta descendo.

Quando $\omega = \omega_r$, o valor será:

$10\log_{10}(4\xi^2)$

Como $xi$ é um número entre 0 e 1, o valor acima dará negativo para valores de $xi$ entre 0 e 0.5. Para valores acima de 0.5, o valor será positivo.

No caso desse termo corresponder a polos, o gráfico será como abaixo. Na aproximação que vamos usar, desenhamos apenas as assíntotas mostrada no gráfico abaixo, em linha cheia grossa.





```python
xi = np.arange(0.2,1.1,0.2)
wr = 2
p = dict()

Happroxz = sym.Piecewise((0, ((omega>=0) & (omega<-polo1))),
                         (-40*sym.log(-omega/polo1,10), (omega>=-polo1)))

p[0] = sym.plot(Happroxz, (omega, 0.1, 100), xscale='log', title=r'$\frac{1}{1-\frac{\omega^2}{\omega_r^2} +j2\xi\frac{\omega}{\omega_r}}$', show=False)



for i in range(0, len(xi)):
    p[i+1]= sym.plot(-10*sym.log((1-omega**2/wr**2)**2+4*xi[i]**2*omega**2/wr**2,10),(omega, 0.1, 40), xscale='log', label=r'$\xi= $'+str(np.round(xi[i],1)), show=False,linestyle='--')
    p[0].extend(p[i+1])



p[0].show()
fig = p[0]._backend.fig
ax = p[0]._backend.ax[0]
ax.set_xticks([0.1, 1, 2, 10])
ax.set_xticklabels([r'$10^{-1}$', r'$10^{0}$', r'$\omega_r$', r'$10^1$']);

for i in range(len(xi)):
    p[0]._backend.ax[0]._children[i+1].set_linestyle('dashed')
p[0]._backend.ax[0]._children[0].set_linewidth(7)
```

#### Fase

Voltamos agora para a expressão da resposta em frequência:

$ H(j\omega) = \frac{Kz_1z_2...z_M}{p_1p_2...p_N}\frac{\left(j\frac{\omega}{z_1} - 1\right)\left(j\frac{\omega}{z_2}-1\right)...\left(j\frac{\omega}{z_M}-1\right)}{\left(j\frac{\omega}{p_1} - 1\right)\left(j\frac{\omega}{p_2}-1\right)...\left(j\frac{\omega}{p_N}-1\right)} $


A fase da resposta em frequência é:

$ \angle{H(j\omega)} = \angle{\frac{Kz_1z_2...z_M}{p_1p_2...p_N}}+\angle{\left(j\frac{\omega}{z_1} - 1\right)}+\angle{\left(j\frac{\omega}{z_2}-1\right)}+...+\angle{\left(j\frac{\omega}{z_M}-1\right)}-\angle{\left(j\frac{\omega}{p_1} - 1\right)}-\angle{\left(j\frac{\omega}{p_2}-1\right)}-...-\angle{\left(j\frac{\omega}{p_N}-1\right)} $

Vamos de novo analisar cada termo, começando do primeiro. Como o termo é um número real, a fase do primeiro termo é zero.

$\angle{\frac{Kz_1z_2...z_M}{p_1p_2...p_N}} = 0$

- **Fase devido a um polo**

Agora vamos analisar um termo correspondente a um polo:

$-\angle{\left(j\frac{\omega}{p_1} - 1\right)} = -\arctan(-\frac{\omega}{p_1})=\arctan(\frac{\omega}{p_1})$

Como o sistema é estável para resposta em frequência existir, o valor do polo é negativo, para $\omega=|p_1|$, a fase será $-45°$. Para $\omega=0$, a fase se 0°. Para $\omega\rightarrow \infty$, a fase se aproxima de $-90°$.

O gráfico da fase devido a um polo  é mostrado abaixo. As assíntotas usadas na aproximação são mostradas em linha grossa contínua. A descida leva 2 décadas começando em $0,1z_1$ e terminando em $10z_1$.

Nesse caso, os termos referentes aos pares complexos conjugados ficam:

Para $\omega=0$, a fase é igual a zero. Para $\omega \rightarrow \infty$, a fase tende a $180°$. Para $\omega=\omega_r$, a fase tende a $90°$.

#### Exemplo 8

Vamos fazer o gráfico da resposta em frequência do sistema com a seguinte função de transferência:

$H(s) = \frac{s+1}{(s+10)(s+30)}$

A resposta em frequência desse sistema é:

$H(j\omega) = \frac{j\omega+1}{(j\omega+10)(j\omega+30)}$

Colocando na forma padrão para construir o diagrama de Bode:

$H(j\omega) = \frac{1}{300}\frac{j\omega+1}{(j\frac{\omega}{10}+1)(j\frac{\omega}{30}+1)}$

Então, em $\omega = 0$ o ganho é $20\log_{10}(1/300)$. Terá uma subida de 20 db/década após 1 rad/s. Após 10 rad/s o gráfico fica plano e após 30 rad/s, a resposta em frequência passa a cair 20 db/década.

Em relação à fase, em $\omega=0$ a fase zero. Em 0,1 rad/s começa a subir. Deveria subir até 10 rad/s (10 vezes o valor do zero do sistema), quando atingiria $\pi/2$. Porém, em 1 rad/s (10 vezes menos do que o polo em -10) o valor fica constante em $\pi/4$. Após 3 rad/s, começa a cair com inclinação $-\pi/4$. Após 10 rad/s passa a cair com o dobro da inclinação, pois a suabida devido ao zero acaba. Após 100 rad/s, a descida passa a ser de $-np.pi/4$. Após 300 rad/s (dez vezes mais que o polo em -30), a fase fica constanta.

O gráfico da resposta em frequência usando a aproximação por retas é mostrado abaixo:

```python
omega = sym.symbols('omega')
Happrox = sym.Piecewise((20*np.log10(1/300), (omega<1)),
                        (20*np.log10(1/300)+20*sym.log(omega,10), (omega>1)&(omega<10)),
                        (20*np.log10(1/300)+20, (omega>=10)&(omega<30)),
                         (20*np.log10(1/300)+20-20*sym.log(omega/30,10), (omega>=30)))

Happroxfase = sym.Piecewise((0, (omega<0.1)),
                           (np.pi/4*sym.log(omega/0.1,10), (omega>=0.1)&(omega<1)),
                           (np.pi/4, (omega>=1)&(omega<3)),
                           (np.pi/4 - np.pi/4*sym.log(omega/3,10), (omega>=3)&(omega<10)),
                           (np.pi/4 - np.pi/4*np.log10(10/3) - np.pi/2*sym.log(omega/10,10), (omega>=10)&(omega<100)),
                           (np.pi/4 - np.pi/4*np.log10(10/3) - np.pi/2 - np.pi/4*sym.log(omega/100,10), (omega>=100)&(omega<300)),
                           (np.pi/4 - np.pi/4*np.log10(10/3) - np.pi/2 - np.pi/4*np.log10(3), (omega>=300)))

p1 = sym.plot(Happrox, (omega, 0.01, 1000), xscale='log', show=False, title=r'$|H(j\omega)|$')
p2 = sym.plot(180/np.pi*Happroxfase, (omega, 0.01, 1000), xscale='log', show=False, title=r'$\angle H(j\omega)$')

PlotGrid(1,2,p1,p2, figsize=(10,3))
```

## Obtendo o gráfico da resposta em frequência com recursos computacionais

A maioria das linguagens usadas em computação científica (como Python, R, Matlab e Julia) tem funções para fazer o gráfico da resposta em frequência de forma exata de um sistema linear e invariante. No Python tem a função `bode` da biblioteca `scipy.signal`. Para usar basta inserir os coeficientes do numerador e do denominador da função de transferência do sistema. Nessa função, o ganho é dado em dB e a fase em graus. Abaixo são obtidos os gráficos do ganho e da fase do  Exemplo 8.

```python
from scipy.signal import bode
import matplotlib.pyplot as plt

num = [1,1]
den = [1,40, 300]
w, ganho, fase = bode((num,den))
plt.figure()
plt.semilogx(w, ganho)
plt.show()
plt.figure()
plt.semilogx(w, fase)
plt.show()
```

## Observações

Os sinais que consideramos aqui foram sinais do tipo:

$x(t) = \cos(\omega t+\theta)$ e $x(t) = e^{j\omega t}$.

Eles têm esse comportamento desde $t=-\infty$. Sinais do tipo:

$x(t) = \cos(\omega t+\theta)u(t)$ e $x(t) = e^{j\omega t}u(t)$

não foram considerados aqui. Quando sinais desse tipo ( que têm seu início em $t=0$) são usados como entrada, a sua saída pode ser dividida em duas partes:

$$y(t) = y_T(t) + y_{RP}(t)$$

$y_T(t)$ corresponde ao que é conhecido como **transitório**. Essa parte fica desprezível depois de algum tempo. O sinal $y_{RP}(t)$ é conhecido como **regime permanente** e é o sinal que calculamos aqui usando a função resposta em frequência. Como os sinais de entrada que consideramos aqui começaram em $t=-\infty$ a parte transitória já sumiu e temos apenas a parte do regime permanente.
