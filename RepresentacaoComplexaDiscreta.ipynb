## Exponencial real

Em sinais discretos a exponencial real é definida da seguinte maneira:

$$x[n] = Ae^{\lambda n}$$

Embora a expressão acima esteja correta, para sinais de tempo discreto é comum definirmos $\alpha = e^\lambda$. Com isso, temos:

$$x[n] = A\alpha^n$$

Com $\alpha$ e $A$ reais.

Note que se $\alpha > 1$, a sequência $x[n]$ sempre aumenta, como é possível ver no gráfico abaixo (com $A=1$ e $\alpha=1.5$):

```python
n = np.arange(20)
A = 1
alpha = 1.5
x = A*alpha**n
plt.stem(n, x);
```

No caso de $0 < \alpha < 1$, a sequência $x[n]$ sempre diminui, como é possível ver no gráfico abaixo (com $A=1$ e $\alpha=0.5$):

```python
n = np.arange(20)
A = 1
alpha = 0.5
x = A*alpha**n
plt.stem(n, x);
```

Para valores de $\alpha$ negativos, o sinal se torna oscilatório (quando $n$ é par, $x[n]$ é positivo, e quando $n$ é ímpar, $x[n]$ é negativo). Isso é possível de se observar no gráfico abaixo (com $A=1$ e $\alpha=-0.8$):

```python
n = np.arange(20)
A = 1
alpha = -0.8
x = A*alpha**n
plt.stem(n, x);
```

O sinal acima oscila e decai. Isso acontece porque $-1<\alpha<0$. Caso $\alpha < -1$, a amplitude do sinal aumentará, como visto abaixo (com $A = 1$ e $\alpha = -1.2$)

```python
n = np.arange(20)
A = 1
alpha = -1.2
x = A*alpha**n
plt.stem(n, x);
```

## Exponencial complexa

Da mesma forma que no caso contínuo, podemos definir um sinal com o expoente complexo:

$$x[n] = e^{j\omega_0 n}$$

Pela relação de Euler-Moivre, este sinal está totalmente relacionado com o cosseno:

$$x[n] = A\cos(\omega_0 n + \theta) = A\frac{e^{j(\omega_0 n + \theta)} + e^{-j(\omega_0 n + \theta)}}{2}$$

Abaixo, são mostrados diversos sinais cosseno, com diferentes frequências (as frequências sobem ao longo dos gráficos, da esquerda para a direita e de cima para baixo).

```python
fig, ax = plt.subplots(3, 4, figsize=(10, 8), sharey=True, sharex=True)


n = np.arange(40)
A = 1
omega0= np.pi/12
x = A*np.cos(omega0*n)
ax[0,0].stem(n, x)
ax[0,0].title.set_text(r'$\omega_0=\pi/12$')

n = np.arange(40)
A = 1
omega0= np.pi/6
x = A*np.cos(omega0*n)
ax[0,1].stem(n, x)
ax[0,1].title.set_text(r'$\omega_0=\pi/6$')

n = np.arange(40)
A = 1
omega0= np.pi/3
x = A*np.cos(omega0*n)
ax[0,2].stem(n, x)
ax[0,2].title.set_text(r'$\omega_0=\pi/3$')

n = np.arange(40)
A = 1
omega0= 2*np.pi/3
x = A*np.cos(omega0*n)
ax[0,3].stem(n, x)
ax[0,3].title.set_text(r'$\omega_0=2\pi/3$')

n = np.arange(40)
A = 1
omega0= np.pi
x = A*np.cos(omega0*n)
ax[1,0].stem(n, x)
ax[1,0].title.set_text(r'$\omega_0=\pi$')

n = np.arange(40)
A = 1
omega0= 4*np.pi/3
x = A*np.cos(omega0*n)
ax[1,1].stem(n, x)
ax[1,1].title.set_text(r'$\omega_0=4\pi/3$')

n = np.arange(40)
A = 1
omega0= 5*np.pi/3
x = A*np.cos(omega0*n)
ax[1,2].stem(n, x)
ax[1,2].title.set_text(r'$\omega_0=5\pi/3$')

n = np.arange(40)
A = 1
omega0= 11*np.pi/6
x = A*np.cos(omega0*n)
ax[1,3].stem(n, x)
ax[1,3].title.set_text(r'$\omega_0=11\pi/6$')

n = np.arange(40)
A = 1
omega0= 23*np.pi/12
x = A*np.cos(omega0*n)
ax[2,0].stem(n, x)
ax[2,0].title.set_text(r'$\omega_0=23\pi/12$')

n = np.arange(40)
A = 1
omega0= 2*np.pi
x = A*np.cos(omega0*n)
ax[2,1].stem(n, x)
ax[2,1].title.set_text(r'$\omega_0=2\pi$')

n = np.arange(40)
A = 1
omega0= 25*np.pi/12
x = A*np.cos(omega0*n)
ax[2,2].stem(n, x)
ax[2,2].title.set_text(r'$\omega_0=25\pi/12$')

n = np.arange(40)
A = 1
omega0= 13*np.pi/6
x = A*np.cos(omega0*n)
ax[2,3].stem(n, x)
ax[2,3].title.set_text(r'$\omega_0=13\pi/6$')
```

Note que até $\omega_0=\pi$ a oscilação ficou mais rápida com o crescimento de $\omega_0$. Porém, diferentemente do sinal de tempo contínuo, que aumenta a oscilação fica sempre mais rápida com o aumento da frequência, no sinal de tempo discreto isso não acontece. Existem duas coisas a serem observadas aqui:

- quando aumentamos a frequência em $2\pi$, o sinal não muda. Isso acontece porque $e^{j(\omega_0+2\pi)n} = e^{j\omega_0 n}e^{j2\pi n}=e^{j\omega_0}1=e^{j\omega_0 n}$. Por isso quando $\omega_0 = \frac{25\pi}{12}$ o sinal é igual a quando $\omega_0 = \frac{\pi}{12}$ ($\frac{\pi}{12} + 2\pi= \frac{25\pi}{12}$). Da mesma forma, quando $\omega_0 = \frac{13\pi}{6}$ o sinal é igual a quando $\omega_0 = \frac{\pi}{6}$ ($\frac{\pi}{6} + 2\pi= \frac{13\pi}{6}$).

- quando $\pi \leq \omega_0 < 2\pi$ (ou esses valores adicionados de múltiplos de $2\pi$), podemos interpretar $\omega_0$ como sendo o negativo do seu complemento para $2\pi$ (por exemplo, para $\omega_0=\frac{5\pi}{3}$, $e^{j\frac{5\pi}{3}n} = e^{-j\frac{\pi}{3}n}$). Então, pela fórmula de Euler-Moivre:

$$\cos\left(\frac{5\pi}{3}n\right) = \frac{e^{j\frac{5\pi}{3}n}+e^{-j\frac{5\pi}{3}n}}{2} = \frac{e^{-j\frac{\pi}{3}n}+e^{j\frac{\pi}{3}n}}{2}=\cos\left(\frac{\pi}{3}n\right) $$

Por isso, o sinal com $\omega_0=\frac{11\pi}{6}$ é igual ao sinal de quando $\omega_0=\frac{\pi}{6}$, o sinal com $\omega_0=\frac{4\pi}{3}$ é igual ao sinal de quando $\omega_0=\frac{2\pi}{3}$ e o sinal com $\omega_0=\frac{23\pi}{12}$ é igual ao sinal de quando $\omega_0=\frac{\pi}{12}$.

Uma outra diferença em relação ao cosseno (e exponencial complexa) de tempo contínuo é que nem sempre o sinal vai ser periódico. Para percebermos isso, vamos usar a definição de sinal periódico:

$$e^{j\omega_0 n} = e^{j\omega_0 (n+N)} = e^{j\omega_0 n}e^{j\omega_0 N}$$

A expressão acima só vai ter os dois lados iguais se:

$$ e^{j\omega_0 N} = 1$$

Isso só vai acontecer se existir algum $N$ inteiro que faça $\omega_0 N=2\pi$. Se existir, esse $N$ é que é o período fundamental do sinal. Mas, para existir, $\frac{\omega_0}{2\pi}$ precisa ser racional. Se essa divisão não for racional, o sinal não será periódico. Por exemplo, para:

$$x[n] = \cos(0.6n)$$

```python
n = np.arange(40)
A = 1
omega0 = 0.6
x = np.cos(omega0*n)
plt.stem(n, x)
```

Note que os valores nunca se repetem (e nunca irão se repetir), e portanto o sinal é não periódico. Podemos notar que $\frac{0.6}{2\pi}$ é não racional (não é posível escrever como fração).

### Caso geral

Um sinal exponencial complexo mais geral é da forma:

$$x[n] = \frac{Ae^{rn + j\omega n + j\theta} + Ae^{rn - j\omega n - j\theta}}{2} = Ae^{rn}\frac{e^{j(\omega n + \theta)}+e^{-j(\omega n + \theta)}}{2} = Ae^{rn}\cos(\omega n+\theta) = A\alpha^n\cos(\omega n+\theta)$$

O gráfico desse sinal, caso $alpha>1$, é a cossenoide com a sua amplitude modulada pela exponencial crescente:

```python
n = np.arange(40)
A = 1
alpha = 1.1
omega0= 11*np.pi/6
x = A*alpha**n*np.cos(omega0*n)
plt.stem(n, x);
```

E no caso de $0<\alpha<1$, é a cossenoide com a sua amplitude modulada pela exponencial decrescente:

```python
n = np.arange(40)
A = 1
alpha = 0.8
omega0= 11*np.pi/6
x = A*alpha**n*np.cos(omega0*n)
plt.stem(n, x);
```

### Somatório de exponencias complexas harmonicamente relacionadas

Exponenciais harmonicamente relacionadas $\phi_k[n]$ são exponenciais complexas que têm um período $N$ em comum (o termo "harmônico" tem o mesmo significado do empregado em música). Ou seja, são sinais com frequências que são múltiplas da menor frequência encontrada entre os sinais.

$\phi_k[n] = A_ke^{jk\omega_0 n}, k = 0,\pm 1,\pm 2 ...$

Para $k = 0$ o sinal é uma constante e para os outros valores de $k$ a frequência angular do sinal  é $|k|\omega_0$. O período fundamental de um sinal que seja a soma dessas exponenciais harmônicas é:

$N_0 = \frac{2\pi k}{\omega_0}$

com $N_0$ inteiro e $k$ inteiro. Se não existir $N_0$ inteiro para nenhum $k$ inteiro, o sinal não é periódico.

Devido à fórmula de Euler-Moivre, isso também vale para conjunto de senos e cossenos que sejam harmonicamente relacionados.

Por exemplo, o sinal:

$x[n] = \cos\left(\frac{\pi}{6}n\right) + 0.5\cos\left(2\frac{\pi}{6}n\right)$

a frequência $\omega_0=\frac{\pi}{6}$ rad. Então o período fundamental do sinal é $N_0=\frac{2\pi}{\pi/6} = 12$, como é possível observar pelo gráfico abaixo ( o gráfico se repete a cada 12 instantes discretos):

```python
n = np.arange(40)
x = np.cos(np.pi/6*n)+0.5*np.cos(np.pi/3*n)
plt.stem(n, x);
```

Nem sempre é tão óbvio qual a frequência angular fundamental $\omega_0$. Por exemplo, o sinal abaixo:

$x[n] = \cos\left(0,07\pi n\right) + 0.5\cos\left(0,06\pi n\right)$

Num primeiro momento você pode achar que $\omega_0=0,06\pi$. Mas $0,07\pi$ não é múltiplo de $0,06\pi$, então esta não é a frequência angular fundamental. Nesse caso $\omega_0= 0,01\pi$ e o período fundamental é $N_0=\frac{2\pi k}{0,01\pi}=\frac{200k}{1} = 200$ (com $k = 1$).

```python
n = np.arange(300)
x = np.cos(0.07*np.pi*n)+np.cos(0.06*np.pi*n)
plt.stem(n, x);
```

Então, de uma forma geral, para encontrarmos a frequência fundamental $\omega_0$ em um sinal gerado por uma soma de senoides é o máximo divisor comum entre as frequências de cada senoide que compõe o sinal.
